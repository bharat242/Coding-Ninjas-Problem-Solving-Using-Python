# Largest Row or Column
For a given two-dimensional integer array/list of size (N x M), you need to find out which row or column has the largest sum(sum of all the elements in a row/column) amongst all the rows and columns.<br>
<br>
Note :<br>
If there are more than one rows/columns with maximum sum, consider the row/column that comes first. And if ith row and jth column has the same largest sum, consider the ith row as answer.<br>
<br>
Input Format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains two integer values, 'N' and 'M', separated by a single space. They represent the 'rows' and 'columns' respectively, for the two-dimensional array/list.<br>
<br>
Second line onwards, the next 'N' lines or rows represent the ith row values.<br>
<br>
Each of the ith row constitutes 'M' column values separated by a single space.<br>
<br>
Output Format :<br>
For each test case, If row sum is maximum, then print: "row" <row_index> <row_sum><br>
OR<br>
If column sum is maximum, then print: "column" <col_index> <col_sum><br>
It will be printed in a single line separated by a single space between each piece of information.<br>
<br>
Output for every test case will be printed in a seperate line.<br>
 Consider :<br>
If there doesn't exist a sum at all then print "row 0 -2147483648", where -2147483648 or -2^31 is the smallest value for the range of Integer.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^3<br>
0 <= M <= 10^3<br>
Time Limit: 1sec<br>
Sample Input 1 :<br>
1<br>
2 2 <br>
1 1 <br>
1 1<br>
Sample Output 1 :<br>
row 0 2<br>
Sample Input 2 :<br>
2<br>
3 3<br>
3 6 9 <br>
1 4 7 <br>
2 8 9<br>
4 2<br>
1 2<br>
90 100<br>
3 40<br>
-10 200<br>
Sample Output 2 :<br>
column 2 25<br>
column 1 342

```python
#Remove All Code From Coding Ninjas then apply
from sys import stdin
MIN_VALUE=-2147483648
def findLargest(arr, nRows, mCols):
    #Your code goes here
    isRow =True
    largestSum=MIN_VALUE
    num=0
    for i in range(nRows):
        rowSum=0
        for j in range(mCols):
            rowSum+=arr[i][j]
        if rowSum>largestSum:
            largestSum=rowSum
            num=i
    for j in range(mCols):
        colSum=0
        for i in range(nRows):
            colSum+=arr[i][j]
        if colSum>largestSum:
            largestSum=colSum
            num=j
            isRow=False
    if isRow:
        print("row "+str(num)+" "+str(largestSum))
    else:
        print("column "+str(num)+" "+str(largestSum))
#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    findLargest(mat, nRows, mCols)

    t -= 1
```
